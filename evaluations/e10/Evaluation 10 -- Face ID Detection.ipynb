###Helpers for visualizing results

```python
import numpy as np
import time
from detect import ShirtRecognizer, FaceRecognizer, get_color
```

```python
def display_image(image):
    ''' Show an image in a matplotlib figure. '''
    figure()  # make a new figure for each image
    pyplot_img = image[:,:,[2,1,0]]  # Convert coloring to matplotlib RGB
    imshow(pyplot_img)
```

```python
def draw_clothing_box(image, face, ratio=.75, heads_below=1.5):
    ''' Draw box around clothing patch. '''
    
    # Get the bounds of the shirt
    shirt_recognizer = ShirtRecognizer()
    shirt = shirt_recognizer.get_shirt(image, face)
    
    if shirt is not None:
        # Inner rectangle shows average color of the clothing patch. '''
        # Outer rectangle, for edge. '''
        left, top, width, height = shirt
        color = get_color(image, shirt)
        cv2.rectangle(image,
            (left, top),
            (left + width, top + height),
            color, -1)
        cv2.rectangle(image,
            (left, top),
            (left + width, top + height),
            (0, 0, 255), 10)
```

```python
def mark_shirt(image):
    ''' Detect and draw all faces found in an image. '''
    
    face_recognizer = FaceRecognizer()
    faces = face_recognizer.find_faces(image)

    for face in faces:
        face = face[:4]
        l, t, w, h = face
        cv2.rectangle(image, 
            (l, t),
            (l + w, t + h), 
            (0, 255, 0), 10)
        color = draw_clothing_box(image, face)
        display_image(image)
```
